<h3>Importing Libraries</h3>

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

<h3>Menampilkan Dataset</h3>
Melakukan Konversi nilai k dan M menjadi ribuan dan jutaan dari string menjadi float

```python
df = pd.read_csv('Dataset1B.csv')

def convert_to_numeric(value):
    if 'M' in value:
        return float(value.replace(' M', '')) * 10**6
    elif 'k' in value:
        return float(value.replace(' k', '')) * 10**3
    else:
        return float(value)

# Mengkonversi nilai string menjadi float
df['installs'] = df['installs'].apply(convert_to_numeric)

display(df)
```

<h3>Penambahan Great dan Poor</h3>
<ul><li>Membuat atribut Great dengan menjumlahkan nilai-nilai dari 4 star ratings dan 5 star ratings</li>
<li>Membuat atribut Poor dengan menjumlahkan nilai-nilai dari 1 star ratings dan 2 star ratings</li>
</ul>

```python
df["great"] = df['5 star ratings'] + df['4 star ratings']
df["poor"] = df['1 star ratings'] + df['2 star ratings']
display(df)
```

<h3>Melihat Deskripsi Data</h3>

```python
summary = df.describe()
display(summary)
```

<h3>Function Untuk Plotting</h3>

```python
def box_plot(df, column, plot_color):
    sns.boxplot(data=df, y=column, color=plot_color)
    plt.title(f'Box Plot of {column}')
    plt.grid(alpha=0.5)
    plt.show()
    plt.close()

# Melihat boxplot tanpa outliers agar lebih mudah terbaca
def removed_outliers_box_plot(df, column, plot_color):
    sns.boxplot(data=df, y=column, color=plot_color, showfliers = False)
    plt.title(f'Box Plot without Outliers of {column}')
    plt.grid(alpha=0.5)
    plt.show()
    plt.close()

def histogram_plot(df, column, num_of_bins, plot_color):
    sns.histplot(data=df, x=column, color=plot_color, bins=num_of_bins)
    plt.title(f'Histogram of {column}')
    plt.grid(alpha=0.5)
    plt.show()
    plt.close()

# Melihat histogram tanpa outliers agar lebih mudah terbaca
def removed_outliers_histogram_plot(df, column, upper_value, lower_value, num_of_bins, plot_color):
    removed_outliers = df[(df[column] <= upper_value) & (df[column] >= lower_value)]
    sns.histplot(removed_outliers[column], color=plot_color, bins=num_of_bins)
    plt.title(f'Histogram without Outliers of {column}')
    plt.grid(alpha=0.5)
    plt.show()
    plt.close()
```

<h2>Average Rating</h2>

```python
# Menampilkan Dataframe Average Rating
averageDf = df[['average rating']]
display(averageDf)
```

<h4>Statistika Kuantitatif</h4>

```python
print("Average Rating")

# Ukuran Pusat Data
avg_rating_mean = averageDf.mean().values[0]    # Rata-rata
avg_rating_median = averageDf.median().values[0] # Median
avg_rating_mode = averageDf.mode().values[0][0]  # Modus

print("Ukuran Pusat:")
print(f"Rata-rata: {avg_rating_mean}")
print(f"Median: {avg_rating_median}")
print(f"Modus: {avg_rating_mode}")

# Ukuran Variasi Data
avg_rating_stdev = averageDf.std().values[0]  # Standar Deviasi

print("\nUkuran Variasi:")
print(f"Standar Deviasi: {avg_rating_stdev}")

# Ukuran Lokasi Data
avg_rating_quartile_0 = averageDf.quantile(q=0).values[0] # Kuartil 0
avg_rating_quartile_1 = averageDf.quantile(q=0.25).values[0] # Kuartil 1
avg_rating_quartile_2 = averageDf.quantile(q=0.5).values[0] # Kuartil 2
avg_rating_quartile_3 = averageDf.quantile(q=0.75).values[0] # Kuartil 3
avg_rating_quartile_4 = averageDf.quantile(q=1).values[0] # Kuartil 4
avg_rating_IQR = np.round(1.5 * (avg_rating_quartile_3-avg_rating_quartile_1),3) # Jangkauan Inter Kuartil, dibulatkan 3 angka belakang koma

print("\nUkuran Lokasi:")
print(f"Quartil 0: {avg_rating_quartile_0}")
print(f"Quartil 1: {avg_rating_quartile_1}")
print(f"Quartil 2: {avg_rating_quartile_2}")
print(f"Quartil 3: {avg_rating_quartile_3}")
print(f"Quartil 4: {avg_rating_quartile_4}")
print(f"Jangkauan Inter Kuartil: {avg_rating_IQR}")


# pencilan atas
upper = avg_rating_quartile_3 + avg_rating_IQR
upper_array = np.array(df['average rating'] > upper)
print(f"Pencilan atas berada di atas skor {upper}")
print(f"Jumlah pencilan atas: {upper_array.sum()}")

# pencilan bawah
lower = avg_rating_quartile_1 - avg_rating_IQR
lower_array = np.array(df['average rating'] < lower)
print(f"Pencilan bawah berada di bawah skor {lower}")
print(f"Jumlah pencilan bawah: {lower_array.sum()}")

print(f"Total pencilan: {upper_array.sum() + lower_array.sum()}")

# menampilkan pencilan berdasarkan index
upper_outliers_index = np.where(upper_array)[0].tolist()
lower_outliers_index = np.where(lower_array)[0].tolist()
outliers_index = upper_outliers_index + lower_outliers_index
outliers = df.iloc[outliers_index]

print("\nData-data yang merupakan outliers:")
display(outliers[['title','average rating']])
```

```python
avg_rating_color = 'palevioletred'
avg_rating_box_plot = box_plot(df, 'average rating', avg_rating_color)
avg_rating_box_plot_without_outliers = removed_outliers_box_plot(df, 'average rating', avg_rating_color)
avg_rating_histogram = histogram_plot(df, 'average rating', 10, avg_rating_color)
avg_rating_histogram_without_outliers = removed_outliers_histogram_plot(df, 'average rating', upper, lower, 10, avg_rating_color)
```

<h2>Average Rating per Kategori</h2>

```python
df_group = df.groupby('category')
categories_mean_by_averagerating = df_group['average rating'].mean()
display(categories_mean_by_averagerating)
```

<h4>Box Plot Keseluruhan Kategori</h4>

```python
fig = plt.figure(figsize =(30, 8))
sns.boxplot(data=df, x="category", y="average rating")
plt.grid()
plt.title("Boxplot of Average Rating grouped by category")
plt.show()
```

<h4>Box Plot Per Kategori</h4>

```python
def box_plot_category(df, column, plot_color, category):
    sns.boxplot(data=df, y=column, color=plot_color)
    plt.title(f'Box Plot of {category}')
    plt.show()
    plt.close()
    
category_colors = ['darkgray', 'dimgrey', 'lightcoral', 'red', 'navajowhite', 'darkorange', 
                   'gold', 'goldenrod', 'turquoise', 'yellow', 'springgreen', 'darkgreen', 'blue', 
                   'cyan', 'darkcyan', 'darkmagenta', 'slateblue']

category_color_dict = dict(zip(categories_mean_by_averagerating.index, category_colors))

for category in categories_mean_by_averagerating.index:
    subset = df[df['category'] == category]
    box_plot_category(subset, 'average rating', category_color_dict[category], category)
```

<h4>Histogram Keseluruhan Kategori</h4>

```python
sns.set_style("whitegrid")

fig = plt.figure(figsize =(25, 10))

_ = sns.histplot(data = df, x = "average rating", hue = "category", bins = 15, palette=category_colors)

plt.title("Histogram of Average Rating grouped by Category")
plt.show()
```

<h4>Histogram Per Kategori</h4>

```python
for name, group in df_group:
    # Get the color corresponding to the category name
    color_index = categories_mean_by_averagerating.index.get_loc(name)
    color = category_colors[color_index]
    
    fig = plt.figure(figsize=(10, 5))
    _ = sns.histplot(data=group, x="average rating", bins=15, color=color)
    plt.title(f"Histogram of {name}")
    plt.show()
    plt.close()
```
