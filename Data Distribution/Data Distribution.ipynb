<h4>Importing Libraries</h4>

```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

<h4>Menampilkan Dataset</h4>

```python
df = pd.read_csv('income.csv')
display(df)
```

<h2>INCOME</h2>


<h3>Menampilkan Data Income</h3>

```python
incomeDf = df[['Income']]
display(incomeDf)

# displaying income
income = df[['Income']]
income_array = income.values
print("Income Array\t:")
print(income_array)
```

<h3>Visualisasi Distribusi Data Income</h3>

```python
plt.hist(income, bins=10)
plt.xlabel("Pendapatan (Ribuan USD)")
plt.ylabel("Jumlah")
plt.title("Sebaran Data Pendapatan")
plt.show()
```

<h3>Statistik Data Income</h3>

```python
# menghitung statistik data income

n_income = len(income_array)
x_bar_income = np.mean(income_array)
s_income = stats.tstd(income_array)[0] # method ini digunakan untuk stdev income

print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
```

<h3>Estimasi Rataan</h3>

<h4>Estimasi 2 Sisi Interval Rataan Income</h4>

```python
# menghitung estimasi 2 sisi interval rataan populasi (income) berdasarkan rataan sampel

alpha = 0.05 # tingkat kepercayaan 95%
# mencari nilai z_alpha_half sehingga P(Z>z_alpha_half) = 1 - alpha/2
z_alpha_half = stats.norm.ppf(1-alpha/2, loc=0, scale=1)

# menghitung standard error berdasarkan formula estimasi
standard_error = z_alpha_half * s_income / np.sqrt(n_income)

# menghitung batas bawah dan batas atas
batas_bawah_2_sisi = x_bar_income - standard_error
batas_atas_2_sisi = x_bar_income + standard_error

print("Estimasi 2 sisi rataan populasi income")
print("--------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah_2_sisi))
print("Batas atas interval estimasi\t: {}".format(batas_atas_2_sisi))
```

<h4>Estimasi 1 Sisi Dari Atas Interval Rataan Income</h4>

```python
# menghitung estimasi 1 sisi dari atas interval rataan populasi (income) berdasarkan rataan sampel

alpha = 0.05
# mencari nilai z_alpha sehingga P(Z>z_alpha) = 1 - alpha
z_alpha = stats.norm.ppf(1-alpha, loc=0, scale=1)

# menghitung batas atas
batas_atas = x_bar_income + z_alpha * s_income / np.sqrt(n_income)

print("Estimasi 1 sisi dari atas rataan populasi income")
print("------------------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

<h4>Estimasi 1 Sisi Dari Bawah Interval Rataan Income</h4>

```python
# menghitung estimasi 1 sisi dari bawah interval rataan populasi (income) berdasarkan rataan sampel

alpha = 0.05
# mencari nilai z_alpha sehingga P(Z>z_alpha) = 1 - alpha
z_alpha = stats.norm.ppf(1-alpha, loc=0, scale=1)

# menghitung batas bawah
batas_bawah = x_bar_income - z_alpha * s_income / np.sqrt(n_income)

print("Estimasi 1 sisi dari bawah rataan populasi income")
print("-------------------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
```

<h3>Estimasi Standar Deviasi</h3>

<h4>Estimasi 2 Sisi Interval Standar Deviasi Income</h4>

```python
# menghitung estimasi 2 sisi interval standar deviasi populasi berdasarkan rataan sampel

alpha = 0.05
v = n_income-1

# mencari nilai c_alpha_half sehingga P(Chi<c_alpha_half) = 1 - alpha/2
c_alpha_half = stats.chi2.ppf(1-alpha/2, v)
# mencari nilai c_alpha_half sehingga P(Chi<c_alpha_half) = alpha/2
c_one_min_alpha_half = stats.chi2.ppf(alpha/2, v)

# menghitung batas bawah dan batas atas
batas_bawah = s_income * np.sqrt(v/c_alpha_half)
batas_atas = s_income * np.sqrt(v/c_one_min_alpha_half)

print("Estimasi 2 sisi st dev populasi income")
print("--------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

<h4>Estimasi 1 Sisi Dari Atas Interval Standar Deviasi Income</h4>

```python
# menghitung estimasi 1 sisi dari atas interval standar deviasi populasi berdasarkan rataan sampel

alpha = 0.05
v = n_income-1

# mencari nilai c_alpha sehingga P(Chi<c_alpha) = alpha
c_one_min_alpha = stats.chi2.ppf(alpha, v)

# menghitung batas atas
batas_atas = s_income * np.sqrt(v/c_one_min_alpha)

print("Estimasi 1 sisi dari atas st dev populasi income")
print("------------------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas atas interval estimasi\t: {}".format(batas_atas))
```

<h4>Estimasi 1 Sisi Dari Bawah Interval Standar Deviasi Income</h4>

```python
# menghitung estimasi 1 sisi dari bawah interval standar deviasi populasi berdasarkan rataan sampel

alpha = 0.05
v = n_income-1

# mencari nilai c_alpha sehingga P(Chi<c_alpha) = 1 - alpha
c_alpha = stats.chi2.ppf(1 - alpha, v)

# menghitung batas atas
batas_bawah = s_income * np.sqrt(v/c_alpha)

print("Estimasi 1 sisi dari bawah st dev populasi income")
print("------------------------------------------------")
print("Total data sampel\t: {}".format(n_income))
print("Rata-rata data sampel\t: {}".format(x_bar_income))
print("Stdev data sampel\t: {}".format(s_income))
print()
print("Tingkat kepercayaan estimasi\t: {}".format(1-alpha))
print("Batas bawah interval estimasi\t: {}".format(batas_bawah))
```

<h5>Statistik Data Nasabah yang Prospektif</h5>

```python
#mencari warga yang memiliki pendapatan >60K USD
income_proportion = []

for cust_income in income_array:
    if cust_income > 60 :
        income_proportion.append(1)
    else:
        income_proportion.append(0)

sample_percentage = sum(income_proportion)/n_income   # menghitung persentase nasabah
s_sample_percentage = stats.tstd(income_proportion)

print("Total data sampel             \t: {}".format(n_income))
print("Total calon nasabah prospektif\t: {}".format(sum(income_proportion)))
print("Persentase nasabah prospektif \t: {}".format(sample_percentage))
print("Stdev data sampel             \t: {}".format(s_sample_percentage))
```

<h5>Estimasi 2 Sisi Interval Rataan Nasabah yang Prospektif</h5>
